```python
import pandas as pd

# 1. Definir nombres de variables según el orden proporcionado
column_names = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race", "sex",
    "capital-gain", "capital-loss", "hours-per-week", "native-country", "class"
]

# 2. Cargar la base de datos con los nombres de variables
file_path = "adult.data"  # Asegúrate de que el archivo está en el mismo directorio
df = pd.read_csv(file_path, names=column_names, na_values="?", skipinitialspace=True)  
# `na_values=" ?"` trata los valores faltantes como NaN
# `skipinitialspace=True` elimina espacios en blanco extra
```

```python
print("\n✅ Variables en la base de datos:")
print(df.columns.tolist())  
```

```python
def describe_dataframe(df):
    summary = {}

    for col in df.columns:
        if df[col].dtype == "object":  # Variables categóricas
            summary[col] = {
                "Tipo": "Categórica",
                "Categorías únicas": df[col].nunique(),
                "Frecuencias": df[col].value_counts(dropna=False).to_dict(),
                "Valores faltantes": df[col].isna().sum()
            }
        else:  # Variables numéricas
            summary[col] = {
                "Tipo": "Numérica",
                "Estadísticas básicas": df[col].describe().to_dict(),
                "Valores faltantes": df[col].isna().sum()
            }

    return summary

descripcion = describe_dataframe(df)

# 5. Mostrar descripción
import pprint
pprint.pprint(descripcion)
```
